```python
# Bar Chart - CO2 Emissions (With green bars)
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("renewable_energy.csv")

emissions_dict = dict(zip(df["Country"], df["CO2_Emissions"]))

plt.bar(emissions_dict.keys(), emissions_dict.values(), color="green")
plt.xticks(rotation=45)
plt.title("CO2 Emissions per Capita by Country")
plt.ylabel("Tonnes")
plt.grid(True)
plt.show()
```

```python
# Scatterplot - Population vs Renewable
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("renewable_energy.csv")

plt.scatter(df["Population"], df["%Renewable"])

for i in range(len(df)):
    plt.text(df["Population"][i], df["%Renewable"][i], df["Country"][i])

plt.title("Population vs % Renewable Energy Use")
plt.xlabel("Population (millions)")
plt.ylabel("% Renewable")
plt.grid(True)
plt.show()
```

```python
# Scatterplot - Population vs Renewable
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("renewable_energy.csv")
df_asia = df[df["Region"] == "Asia"].reset_index() # Reset index needed to enable iteration

plt.scatter(df_asia["Population"], df_asia["%Renewable"])

for i in range(len(df_asia)):
    plt.text(df_asia["Population"][i], df_asia["%Renewable"][i], df_asia["Country"][i])

plt.title("Population vs % Renewable Energy Use")
plt.xlabel("Population (millions)")
plt.ylabel("% Renewable")
plt.grid(True)
plt.show()
```
